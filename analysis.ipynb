## RFM Analysis

```python
# Snapshot date (1 day after last order)
snapshot_date = df['order_date'].max() + pd.Timedelta(days=1)

rfm = (
    df.groupby('customer_id')
      .agg(
          recency=('order_date', lambda x: (snapshot_date - x.max()).days),
          frequency=('order_id', 'count'),
          monetary=('order_value', 'sum')
      )
      .reset_index()
)
```

```python
# RFM scores using rank to avoid duplicate bin edges
rfm['R'] = pd.qcut(rfm['recency'].rank(method='first'), 5, labels=[5,4,3,2,1]).astype(int)
rfm['F'] = pd.qcut(rfm['frequency'].rank(method='first'), 5, labels=[1,2,3,4,5]).astype(int)
rfm['M'] = pd.qcut(rfm['monetary'].rank(method='first'), 5, labels=[1,2,3,4,5]).astype(int)
```

```python
def rfm_segment(row):
    R, F, M = row['R'], row['F'], row['M']

    if R == 5 and F == 5 and M == 5:
        return 'Top Customers'

    elif F >= 4 and M >= 4:
        return 'Loyal Customers'

    elif R == 5:
        return 'Recent Customers'

    elif M == 5:
        return 'Big Spenders'

    elif R == 1:
        return 'Lost Customers'

    elif R == 1 and F == 1 and M == 1:
        return 'Low Value Customers'

    elif R >= 4 and F >= 3:
        return 'Potential Loyalists'

    elif R in [2,3] and F in [2,3]:
        return 'At Risk Customers'

    else:
        return 'Others'
```

```python
rfm['segment'] = rfm.apply(rfm_segment, axis=1)
```

```python
rfm_segments = (
    rfm.groupby('segment', as_index=False)
       .agg(
           customers=('customer_id', 'count'),
           avg_recency=('recency', 'mean'),
           avg_frequency=('frequency', 'mean'),
           avg_monetary=('monetary', 'mean')
       )
)

rfm_segments
```

```python
rfm_ops = (
    df.merge(rfm[['customer_id', 'segment']], on='customer_id')
      .groupby('segment', as_index=False)
      .agg(
          total_orders=('order_id', 'count'),
          complaints=('complaint_flag', 'sum'),
          returned_orders=('return_status', lambda x: (x == 'Returned').sum())
      )
)

rfm_ops['complaint_rate'] = rfm_ops['complaints'] / rfm_ops['total_orders']
rfm_ops['return_rate'] = rfm_ops['returned_orders'] / rfm_ops['total_orders']

rfm_ops
```

## BCG Matrix (Category)

```python
# Ensure order_month exists
df['order_month'] = pd.to_datetime(df['order_date']).dt.to_period('M').astype(str)

category_monthly = (
    df.groupby(['order_month', 'product_category'], as_index=False)
      .agg(revenue=('order_value', 'sum'))
)
```

```python
# Pivot for growth calculation
category_pivot = category_monthly.pivot(
    index='order_month',
    columns='product_category',
    values='revenue'
).sort_index()

# Growth = (last month - first month) / first month
category_growth = (
    (category_pivot.iloc[-1] - category_pivot.iloc[0]) /
     category_pivot.iloc[0]
).reset_index()

category_growth.columns = ['product_category', 'revenue_growth']
```

```python
category_revenue = (
    df.groupby('product_category', as_index=False)
      .agg(total_revenue=('order_value', 'sum'))
)

category_revenue['revenue_share'] = (
    category_revenue['total_revenue'] /
    category_revenue['total_revenue'].sum()
)
```

```python
bcg_category = category_revenue.merge(
    category_growth,
    on='product_category',
    how='left'
)

bcg_category
```

```python
# Thresholds (medians used for internal BCG)
share_median = bcg_category['revenue_share'].median()
growth_median = bcg_category['revenue_growth'].median()

def bcg_quadrant(row):
    if row['revenue_share'] >= share_median and row['revenue_growth'] >= growth_median:
        return 'Star'
    elif row['revenue_share'] >= share_median and row['revenue_growth'] < growth_median:
        return 'Cash Cow'
    elif row['revenue_share'] < share_median and row['revenue_growth'] >= growth_median:
        return 'Question Mark'
    else:
        return 'Dog'

bcg_category['bcg_quadrant'] = bcg_category.apply(bcg_quadrant, axis=1)

bcg_category[['product_category', 'revenue_share', 'revenue_growth', 'bcg_quadrant']] \
    .sort_values(['bcg_quadrant', 'revenue_share'], ascending=[True, False])
```
